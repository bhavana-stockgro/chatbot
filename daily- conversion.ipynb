```python
#importing necessary libraries
from datetime import datetime, timedelta
import pandas as pd
import numpy as np
import itertools
```

```python
df = pd.read_csv('BANKNIFTY_minute_wise_19_22.csv')
df['Date'] = pd.to_datetime(df['Date'], format='%Y%m%d').dt.strftime('%Y-%m-%d')
df['Time'] = pd.to_datetime(df['Time']).dt.time
df = df.sort_values(['Date','Time'])
df
```

```python
df = df.drop(['Volume','OI'], axis = 1)
#df
```

```python
df[df['Date']=='2019-04-01']
```

```python
dates = df['Date'].unique()
#dates
```

```python
final_df = pd.DataFrame(columns = ['Date','Time','Open','High','Low','Close']) 
```

```python
new_row_count = df.shape[0] // 10
```

```python
for date in dates:
    date_data = df[df['Date'] == date]
    num_rows = len(date_data)
    for i in range(0, num_rows, 60):
        start_index = i
        end_index = min(i + 10, num_rows)
        averaged_open = date_data['Open'][start_index:end_index].mean()
        final_df = final_df.append({
            "Date": date,
            "Time": date_data.iat[start_index, date_data.columns.get_loc("Time")],
            "Open": averaged_open,
            "High": date_data.iat[start_index, date_data.columns.get_loc("High")],
            "Low": date_data.iat[start_index, date_data.columns.get_loc("Low")],
            "Close": date_data.iat[start_index, date_data.columns.get_loc("Close")]
        }, ignore_index=True)
```

```python
final_df.head(50)
```

```python
final_df.to_csv('concise_BANKNIFTY_minute_wise_19_22.csv', index=False)
```
