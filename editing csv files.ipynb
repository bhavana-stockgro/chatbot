```python
def convert_to_date_format(number):
    number = str(number)
    year = number[:4]
    month = number[4:6]
    day = number[6:]
    formatted_date = f"{year}/{month}/{day}"
    return formatted_date
```

```python
df2['Date'] = df2['Date'].apply(convert_to_date_format)
df2
```

```python
df2['Date'] = pd.to_datetime(df2['Date'],errors='coerce').dt.date
```
